```python
import pandas
```

```python
liberec = pandas.read_csv("zam_liberec.csv")
plzen = pandas.read_csv("zam_plzeň.csv")
praha = pandas.read_csv("zam_praha.csv")
liberec["mesto"] = "Liberec"
plzen["mesto"] = "Plzeň"
praha["mesto"] = "Praha"
zamestnanci = pandas.concat([liberec, plzen, praha], ignore_index=True)
# zamestnanci.head()
zamestnanci.shape
```

```python
platy = pandas.read_csv("platy_2021_02.csv")
platy.head()
```

```python
platy_zamestnancu = pandas.merge(zamestnanci, platy, on="cislo_zamestnance", how="left")
# platy_zamestnancu.shape
platy_zamestnancu.head()
```

```python
aktualni_zamestnanci = platy_zamestnancu[platy_zamestnancu["plat"].notnull()]
aktualni_zamestnanci.head()
```

```python
aktualni_zamestnanci.groupby("mesto")["plat"].mean()
```

```python
vykazy = pandas.read_csv("vykazy.csv")
vykazy.head()
```

```python
vykazy.groupby("project")["hours"].count()
```

```python
byvaly_zamestnanci = platy_zamestnancu[platy_zamestnancu["plat"].isnull()]
byvaly_zamestnanci.to_csv("byvaly_zamestnanci",index=False)
```

```python
len(byvaly_zamestnanci)
f"Počet bývalých zaměstnanců:{len(byvaly_zamestnanci)}"
```

```python
celek = pandas.merge(aktualni_zamestnanci, novy_vykazy, on="cislo_zamestnance")
celek
```
